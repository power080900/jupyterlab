```python
https://comic.naver.com/genre/bestChallenge.nhn 페이지의 웹툰정보를 추출해서 
title, summary  그리고 rating 정보를 각각 화면에 출력한다. 
```

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
import time

titles = []
summarypkg = []
ratings = []
for i in range(10) :    
    urlstr = ('https://comic.naver.com/genre/bestChallenge?&page='+str(i))
    r = requests.get(urlstr)
    html = r.text
    bs= BeautifulSoup(html, 'html.parser') 
    title = bs.select('.challengeTitle > a')
    for tt in title:
        titles.append(tt.string.strip())
    summary = bs.select('.challengeInfo > .summary')
    for sm in summary:
        summarypkg.append(sm.string)
    rating = bs.select('.rating_type > strong')
    for rt in rating:
        ratings.append(rt.string)
    time.sleep(5)
```

```python
print("[웹툰제목]")
print(titles)
print("")
print("[웹툰소개]")
print(summarypkg)
print("")
print("[웹툰평점]")
print(ratings)    
```
