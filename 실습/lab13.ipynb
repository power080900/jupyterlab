```python
import pandas as pd

data = {
    '1분기':[500, 690, 1100, 1500, 1990, 1020],
    '2분기':[450, 700, 1030, 1650, 2020, 1600],
    '3분기':[520, 820, 1200, 1700, 2300, 2200],
    '4분기':[610, 900, 1380, 1850, 2420, 2550],
}
df1 = pd.DataFrame(data)
df1.index = (2015,2016,2017,2018,2019,2020)
df1.index.name = "year"
df1.columns.name = "quarter"
print(df1)
```

```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc  
if sys.platform  == 'darwin':                             ## MAC OS의 이름을 확인한다.
    path = '.....'  
elif sys.platform == 'win32':                             ## Windows 이름을 확인한다.
    path = "font/MaplestoryBold.ttf"
else:
    print('Unknown system... sorry~~~~') 
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False 
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
labels1 = ["1분기","2분기","3분기","4분기"]

plt.figure(figsize=(10,6))
plt.plot(df1, label=labels1)
plt.title('시각화 과제1', color ='purple', size= 20, fontweight = 'bold')
plt.legend()
plt.savefig("output/hw1.png")
```

```python
df1.T.plot.line(figsize=(10,6),marker='s')
plt.savefig("output/hw2.png")
```

```python
labels1 = ["1분기","2분기","3분기","4분기"]

plt.figure(figsize=(10,6))
plt.pie(df1.sum(), labels=labels1,autopct='%.1f%%')
plt.savefig("output/hw3.png")
```

```python
labels2 = ["2015년","2016년","2017년","2018년","2019년","2020년"]

plt.figure(figsize=(10,6))
plt.pie(df1.T.sum(), labels=labels2,autopct='%.1f%%')
plt.savefig("output/hw4.png")
```
