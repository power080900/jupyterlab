```python
df = pd.read_csv('./data/product_click.log',
                 header=None,
                 sep=" ",
                 dtype="object")
df.columns = ['oldclicktime','clickpid']
df['newclicktime'] = pd.to_datetime(df['oldclicktime'])
print("2-1.")
print(df.info())
print(df.head(5))
```

```python
df_time = df
df_time['year'] = df_time['newclicktime'].dt.year
df_time['month'] = df_time['newclicktime'].dt.month
df_time['day'] = df_time['newclicktime'].dt.day
df_time['hour'] = df_time['newclicktime'].dt.hour
df_time['minute'] = df_time['newclicktime'].dt.minute
print("2-2.")
print(df_time.info())
print(df_time.head())
```

```python
df2 = pd.read_csv('./data/cctv_seoul.csv')

def ratio_two_obj(a, b):    # 두 객체의 합
    return a / b * 100

df2['CCTV비율'] = df2.apply(lambda x: ratio_two_obj(x['CCTV수'], x['인구수']), axis=1)   
df2['노인비율'] = df2.apply(lambda x: ratio_two_obj(x['고령자'], x['인구수']), axis=1)   

print('3')
print(df2.head())
```
